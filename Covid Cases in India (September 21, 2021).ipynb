```r
install.packages('tidyverse')
install.packages("gridExtra")
```

```r
data <- read.csv("~/R Projects/Covid_India.csv")
data
```

```r
str(data)
```

```r
library(tidyverse)
```

```r
library(ggplot2)
ggplot <- library(ggplot2)
```

```r
ggplot(data, aes(x = Total.Cases)) + labs(x = 'Total Number of Cases', y='Count') + geom_histogram(fill="royalblue", color="white", alpha=0.9) + theme_minimal()
```

```r
ggplot(data, aes(State.UTs, Total.Cases)) + geom_bar(stat="identity", fill="royalblue") + labs(x = 'Names of Indian States and Union Territories', y='Total Number of Cases') + coord_flip()
```

```r
ggplot(data=data, aes(State.UTs, y=Deaths)) + geom_bar(stat='identity', fill="royalblue") + labs(x='Names of Indian States and Union Territories', y='Number of Deaths') + coord_flip()
```

#### Let's compare the two and remodel the charts

```r
p1 <- ggplot(data=data, aes(x=reorder(State.UTs, Total.Cases), y=Total.Cases)) + geom_bar(stat='identity', fill = 'tomato3') + coord_flip() + labs(x='Names of Indian States and Union Territories', y='Total Number of Cases') + theme_minimal()
p1
```

```r
p2 <- ggplot(data=data, aes(x=reorder(State.UTs, Deaths), y=Deaths)) + geom_bar(stat='identity', fill = 'tomato3') + coord_flip() + labs(x = 'Names of Indian States and Union Territories', y = 'Number of Deaths') + theme_minimal()
p2
```

#### Maharashtra is noticeably greater in the number of cases and deaths

```r
ggplot(data=data, aes(x=Total.Cases, y=Deaths)) + geom_jitter(aes(color=State.UTs), size = 3) + labs(x='Total Number of Cases', y = 'Number of deaths')
library(repr)
options(repr.plot.width=10, repr.plot.height=8)
```

```r
ggplot(data=data, aes(x=Active.Ratio...., y=Discharge.Ratio....)) + geom_jitter(aes(color=State.UTs), size = 3) + labs(x='Active Cases Ratio', y = 'Discharged Cases Ratio')
options(repr.plot.width=10, repr.plot.height=8)
```

#### Maharashtra is an outlier in both instances

```r
ggplot(data=data, aes(x=reorder(State.UTs, Active), y=Active)) + geom_bar(stat='identity', fill = 'tomato3') + coord_flip() + labs(x = 'Names of Indian States and Union Territories', y = 'Active Cases') + theme_minimal()
```

```r
ggplot(data=data, aes(x=reorder(State.UTs, Discharged), y=Discharged)) + geom_bar(stat='identity', fill = 'tomato3') + coord_flip() + labs(x = 'Names of Indian States and Union Territories', y = 'Discharged Cases') + theme_minimal()
```
